# 13. 함수 적용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col= '지원번호')
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

```python
df['키'] = df['키'] +'cm' # int라서 오류가 난다. 밑에 해결방법 있음
df
```

## 데이터에 함수 적용(apply)

```python
# 키 뒤에 cm 를 붙이는 역할
def add_cm(height):
    return str(height) +'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN 이 아닌지
        return lang.capitalize() # 첫글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
df['SW특기'].str.capitalize() # 그냥 이걸 사용하는 것이 편하긴 하다.
#위에 함수는 공부하기 위해서 한 것이다.
```

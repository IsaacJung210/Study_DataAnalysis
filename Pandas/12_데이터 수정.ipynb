# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col= '지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고'},inplace=True)
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
90 + 85 + 100 + 95 + 85
```

```python
df['결과'] = 'Fail' # 결과 Column 을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 Pass 로 업데이트
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 Column을 삭제
```

```python
df.drop(columns=['국어','영어','수학']) # 국어, 영어, 수학 Column을 삭제 , inplace=True 해야 저장됨
```

```python
df
```
